```python
def cleanRawText(text):
    text = text.replace('\n','')
    text = text.replace('\n', '')
    return text
```

```python
def get_body_contents(html):
    soup = BeautifulSoup(html, "html5lib")
    for attr in ['head','html','body']:
        if hasattr(soup, attr):
            getattr(soup, attr).unwrap()
    return soup

    
html = "<p>Hello World</p>"
print(get_body_contents(html))
```

```python
from io import StringIO
from lxml import etree

f = StringIO('<foo><bar><x1>hello</x1><x1>world</x1></bar></foo>')
tree = etree.parse(f)
find_text = etree.XPath("//text()")

# and print out the required data
print([(text,tree.getpath( text.getparent())) for text in find_text(tree)])
```
